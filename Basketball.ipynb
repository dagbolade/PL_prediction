```python
import pandas as pd
from IPython.display import display
# Loading the team data
team_data = pd.read_csv('team_data.csv')
# Loading player data
player_data = pd.read_csv('players_data.csv')
# Loading team stats
team_stats = pd.read_csv('team_stats.csv')

# Displaying the first few rows of each DataFrame to verify the data
print('Team Data:')
display(team_data.head())
print('\n')
print('Player Data:')
display(player_data.head())
print('\n')
print('Team stats:')
display(team_stats.head())
```

```python
# Remove the first cilumn from team_data DataFrame
team_data = team_data.iloc[:, 1:]

# Remove the first column from player_data DataFrame
player_data = player_data.iloc[:, 1:]

# Remove the first column from player_stat DataFrame
team_stats = team_stats.iloc[:, 1:]

# Print the updated first few rows of each DataFrame
print('Updated Team Data:')
display(team_data.head())
print('\n')
print('Updated Player Data:')
display(player_data.head())
print('\n')
print('Updated Team stats:')
display(team_stats.head())
```

# Data Preprocessing

```python
# checking for missing values
team_data.isnull().sum()
team_stats.isnull().sum()
player_data.isnull().sum()
```

```python
# checking for duplicates
team_data.duplicated().sum()
team_stats.duplicated().sum()
player_data.duplicated().sum()
```

Investigation

```python
# Identifying the rows in the player-data DataFrame
duplicate_rows = player_data[player_data.duplicated()]

# Print the duplicate rows
print('Duplicate rows in player_data:')
display(duplicate_rows)
```

```python
player_data = player_data.drop_duplicates()

#  verification of the removal of duplicates
print("Number of duplicates in the player_data after removal:", player_data.duplicated().sum())
```

```python
# Checking for outliers
import matplotlib.pyplot as plt

# Player Data
print('Player Data:')
player_columns = ['Points', 'Assists', 'Minutes', 'Field Goals Made', 'Rebounds', 'Steals', 'Blocks', 'Turnovers']
player_data[player_columns].boxplot(figsize=(12, 8))
plt.title('Player Data Boxplots')
plt.xticks(rotation=45)
plt.show()
```

## Performing Exploratory Data Analysis

```python
# Get descriptive statistics for each DataFrame
print("Team Data Statistics:")
display(team_data.describe())

print("\nPlayer Data Statistics:")
display(player_data.describe())

print("\nAdditional Team Data Statistics:")
display(team_stats.describe())
```

```python
# Correlation matrix for team data
team_corr = team_data.corr()
print("Team Data Correlation Matrix:")
print(team_corr)

# Correlation matrix for player data
player_corr = player_data.corr()
print("\nPlayer Data Correlation Matrix:")
print(player_corr)

# Correlation matrix for additional team data
additional_team_corr = team_stats.corr()
print("\nAdditional Team Data Correlation Matrix:")
print(additional_team_corr)
```
