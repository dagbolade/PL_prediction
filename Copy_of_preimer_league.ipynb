```python
# get the current position and calculate points for all teams
def calculate_points_goals(row):
    home_points, away_points = 0, 0
    home_goal_diff = row['FTHG'] - row['FTAG']
    away_goal_diff = row['FTAG'] - row['FTHG']

    if row['FTR'] == 'H':
        home_points = 3
    elif row['FTR'] == 'A':
        away_points = 3
    else:
        home_points = away_points = 1

    return pd.Series([home_points, away_points, home_goal_diff, away_goal_diff, row['FTHG'], row['FTAG']])

# Applying the function to the dataset
df_new[['HomePoints', 'AwayPoints', 'HomeGoalDiff', 'AwayGoalDiff', 'HomeGoals', 'AwayGoals']] = df_new.apply(calculate_points_goals, axis=1)

# Summarizing the data for each team
team_stats = pd.DataFrame(index=teams)

# Calculating total points, goal difference, and goals scored for each team
team_stats['Points'] = df_new.groupby('HomeTeam')['HomePoints'].sum() + df_new.groupby('AwayTeam')['AwayPoints'].sum()
team_stats['GoalDiff'] = df_new.groupby('HomeTeam')['HomeGoalDiff'].sum() + df_new.groupby('AwayTeam')['AwayGoalDiff'].sum()
team_stats['GoalsScored'] = df_new.groupby('HomeTeam')['HomeGoals'].sum() + df_new.groupby('AwayTeam')['AwayGoals'].sum()

# Sorting the teams based on Points, Goal Difference, and Goals Scored
sorted_teams = team_stats.sort_values(by=['Points', 'GoalDiff', 'GoalsScored'], ascending=[False, False, False])

print('Current standings:')
print(sorted_teams)
```

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Encoding Team Names
label_encoder = LabelEncoder()
df_new['HomeTeamEncoded'] = label_encoder.fit_transform(df_new['HomeTeam'])
df_new['AwayTeamEncoded'] = label_encoder.transform(df_new['AwayTeam'])

# Calculate Team's Average Goals Scored and Conceded per Match
average_goals_scored_home = df_new.groupby('HomeTeam')['FTHG'].mean()
average_goals_conceded_home = df_new.groupby('HomeTeam')['FTAG'].mean()
average_goals_scored_away = df_new.groupby('AwayTeam')['FTAG'].mean()
average_goals_conceded_away = df_new.groupby('AwayTeam')['FTHG'].mean()

# Mapping these averages to the main DataFrame
df_new['HomeTeamAvgGoalsScored'] = df_new['HomeTeam'].map(average_goals_scored_home)
df_new['HomeTeamAvgGoalsConceded'] = df_new['HomeTeam'].map(average_goals_conceded_home)
df_new['AwayTeamAvgGoalsScored'] = df_new['AwayTeam'].map(average_goals_scored_away)
df_new['AwayTeamAvgGoalsConceded'] = df_new['AwayTeam'].map(average_goals_conceded_away)

# Team Form and Performance
# Calculate points earned in the last 5 matches for each team
def calculate_recent_form(team, date, num_matches):
    recent_matches = df_new[(df_new['HomeTeam'] == team) | (df_new['AwayTeam'] == team)]
    recent_matches = recent_matches[recent_matches['Date'] < date].sort_values('Date', ascending=False).head(num_matches)

    points = 0
    for _, row in recent_matches.iterrows():
        if row['HomeTeam'] == team:
            if row['FTR'] == 'H':
                points += 3
            elif row['FTR'] == 'D':
                points += 1
        else:
            if row['FTR'] == 'A':
                points += 3
            elif row['FTR'] == 'D':
                points += 1

    return points

df_new['HomeTeamRecentForm'] = df_new.apply(lambda x: calculate_recent_form(x['HomeTeam'], x['Date'], 5), axis=1)
df_new['AwayTeamRecentForm'] = df_new.apply(lambda x: calculate_recent_form(x['AwayTeam'], x['Date'], 5), axis=1)

# Home/Away Advantage
df_new['HomeAdvantage'] = df_new['FTR'].apply(lambda x: 1 if x == 'H' else 0)

# Player Injuries and Suspensions
# You would need additional data for player injuries and suspensions to incorporate this feature

# Tactical Matchups and Playing Styles
# This would require more advanced analysis and data on team tactics and playing styles

# Expected Goals (xG) and Other Advanced Metrics
# You would need expected goals (xG) data for each team to incorporate this feature

# Viewing the modified DataFrame
display(df_new.head())
```

```python
def calculate_win_percentage(df, team_column, result_column):
    team_wins = df[df[result_column] == 'H'].groupby(team_column).size()
    total_matches = df.groupby(team_column).size()
    team_win_percentage = team_wins / total_matches
    return team_win_percentage

# Calculate Home Team Win Percentage
home_team_win_percentage = calculate_win_percentage(df_new, 'HomeTeam', 'FTR')

# Calculate Away Team Win Percentage
away_team_win_percentage = calculate_win_percentage(df_new, 'AwayTeam', 'FTR')

# Create mappings for win percentages
home_team_win_percentage_map = home_team_win_percentage.to_dict()
away_team_win_percentage_map = away_team_win_percentage.to_dict()

# Map the win percentages to the original DataFrame
df_new['HomeTeamWinPercentage'] = df_new['HomeTeam'].map(home_team_win_percentage_map).fillna(0)
df_new['AwayTeamWinPercentage'] = df_new['AwayTeam'].map(away_team_win_percentage_map).fillna(0)

# Check the first few rows to confirm the new features
print(df_new[['HomeTeam', 'HomeTeamWinPercentage', 'AwayTeam', 'AwayTeamWinPercentage']].head())
```

```python
def calculate_points(df, team_column, result_column):
    points = df[result_column].apply(lambda x: 3 if x == 'H' else 1 if x == 'D' else 0)
    return df.groupby(team_column).apply(lambda x: points[x.index].cumsum()).reset_index(drop=True)

def calculate_recent_form(df, team_column, points_column, window=5):
    form = df.groupby(team_column).apply(lambda x: x[points_column].diff().fillna(0).rolling(window=window+1, min_periods=1).sum().shift(fill_value=0)).reset_index(drop=True)
    return form

# Ensure the DataFrame is sorted by date
df_new.sort_values('Date', inplace=True)
df_new.reset_index(inplace=True)

# Calculate points for each team
df_new['HomeTeamPoints'] = calculate_points(df_new, 'HomeTeam', 'FTR')
df_new['AwayTeamPoints'] = calculate_points(df_new, 'AwayTeam', 'FTR')

# Calculate recent form for each team
df_new['HomeTeamRecentForm'] = calculate_recent_form(df_new, 'HomeTeam', 'HomeTeamPoints')
df_new['AwayTeamRecentForm'] = calculate_recent_form(df_new, 'AwayTeam', 'AwayTeamPoints')

# Football-specific suggestions:
# 1. Calculate goal difference for each team
df_new['HomeTeamGoalDifference'] = df_new.groupby('HomeTeam').apply(lambda x: (x['FTHG'] - x['FTAG']).cumsum()).reset_index(drop=True)
df_new['AwayTeamGoalDifference'] = df_new.groupby('AwayTeam').apply(lambda x: (x['FTAG'] - x['FTHG']).cumsum()).reset_index(drop=True)

# 2. Calculate the head-to-head record between teams
def head_to_head(df, home_team, away_team):
    matches = df[((df['HomeTeam'] == home_team) & (df['AwayTeam'] == away_team)) | ((df['HomeTeam'] == away_team) & (df['AwayTeam'] == home_team))]
    home_wins = matches[matches['FTR'] == 'H'].shape[0]
    away_wins = matches[matches['FTR'] == 'A'].shape[0]
    draws = matches[matches['FTR'] == 'D'].shape[0]
    return pd.Series([home_wins, away_wins, draws], index=['HomeTeamHeadToHeadWins', 'AwayTeamHeadToHeadWins', 'HeadToHeadDraws'])

head_to_head_records = df_new.apply(lambda x: head_to_head(df_new, x['HomeTeam'], x['AwayTeam']), axis=1)

# 3. Calculate the average goals scored and conceded by each team
df_new['HomeTeamAvgGoalsScored'] = df_new.groupby('HomeTeam')['FTHG'].expanding().mean().reset_index(level=0, drop=True)
df_new['AwayTeamAvgGoalsScored'] = df_new.groupby('AwayTeam')['FTAG'].expanding().mean().reset_index(level=0, drop=True)
df_new['HomeTeamAvgGoalsConceded'] = df_new.groupby('HomeTeam')['FTAG'].expanding().mean().reset_index(level=0, drop=True)
df_new['AwayTeamAvgGoalsConceded'] = df_new.groupby('AwayTeam')['FTHG'].expanding().mean().reset_index(level=0, drop=True)

# Merge the remaining features with the main DataFrame
df_new = pd.concat([df_new, head_to_head_records], axis=1)

# Drop the 'index' column
df_new.drop('index', axis=1, inplace=True)

# Display the head of the dataframe to verify
df_new[['Date', 'HomeTeam', 'HomeTeamPoints', 'HomeTeamRecentForm', 'AwayTeam', 'AwayTeamPoints', 'AwayTeamRecentForm',
              'HomeTeamGoalDifference', 'AwayTeamGoalDifference', 'HomeTeamHeadToHeadWins', 'AwayTeamHeadToHeadWins',
              'HeadToHeadDraws', 'HomeTeamAvgGoalsScored', 'AwayTeamAvgGoalsScored', 'HomeTeamAvgGoalsConceded',
              'AwayTeamAvgGoalsConceded']].tail(10)
```

```python
def compute_last_10_games_performance(df):
    df['HomeTeamLast10Wins'] = 0
    df['HomeTeamLast10Draws'] = 0
    df['HomeTeamLast10Losses'] = 0
    df['AwayTeamLast10Wins'] = 0
    df['AwayTeamLast10Draws'] = 0
    df['AwayTeamLast10Losses'] = 0

    teams = df['HomeTeam'].unique()

    for team in teams:
        home_games = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)].sort_values(by='Date')
        home_wins, home_draws, home_losses = [], [], []

        for idx, row in home_games.iterrows():
            if row['HomeTeam'] == team:
                if row['FTR'] == 'H':
                    home_wins.append(1)
                    home_draws.append(0)
                    home_losses.append(0)
                elif row['FTR'] == 'D':
                    home_wins.append(0)
                    home_draws.append(1)
                    home_losses.append(0)
                else:
                    home_wins.append(0)
                    home_draws.append(0)
                    home_losses.append(1)
            else:
                if row['FTR'] == 'A':
                    home_wins.append(1)
                    home_draws.append(0)
                    home_losses.append(0)
                elif row['FTR'] == 'D':
                    home_wins.append(0)
                    home_draws.append(1)
                    home_losses.append(0)
                else:
                    home_wins.append(0)
                    home_draws.append(0)
                    home_losses.append(1)

            if len(home_wins) > 10:
                home_wins.pop(0)
                home_draws.pop(0)
                home_losses.pop(0)

            df.at[idx, 'HomeTeamLast10Wins'] = sum(home_wins[-10:])
            df.at[idx, 'HomeTeamLast10Draws'] = sum(home_draws[-10:])
            df.at[idx, 'HomeTeamLast10Losses'] = sum(home_losses[-10:])
            df.at[idx, 'AwayTeamLast10Wins'] = sum(home_wins[-10:])
            df.at[idx, 'AwayTeamLast10Draws'] = sum(home_draws[-10:])
            df.at[idx, 'AwayTeamLast10Losses'] = sum(home_losses[-10:])

    return df

df_new = compute_last_10_games_performance(df_new)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
import warnings
import pickle
from sklearn.exceptions import ConvergenceWarning

warnings.filterwarnings("ignore", category=ConvergenceWarning)

# Prepare the data
df_new = compute_last_10_games_performance(df_new)  # Ensure the new features are included

X = df_new[['HomeTeamPoints', 'AwayTeamPoints', 'HomeTeamRecentForm', 'AwayTeamRecentForm',
            'HomeTeamGoalDifference', 'AwayTeamGoalDifference', 'HomeTeamAvgGoalsScored',
            'AwayTeamAvgGoalsScored', 'HomeTeamAvgGoalsConceded', 'AwayTeamAvgGoalsConceded',
            'HomeTeamLast10Wins', 'HomeTeamLast10Draws', 'HomeTeamLast10Losses',
            'AwayTeamLast10Wins', 'AwayTeamLast10Draws', 'AwayTeamLast10Losses']]
y = df_new['FTR']

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_resampled)
X_test_scaled = scaler.transform(X_test)

# Define the models
models = [
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),
    ('SVM', SVC(random_state=42)),
    ('Gaussian Naive Bayes', GaussianNB()),
    ('k-NN', KNeighborsClassifier()),
    ('Neural Network (MLP)', MLPClassifier(random_state=42)),
    ('XGBoost', XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')),
    ('LightGBM', LGBMClassifier(random_state=42, verbose=-1)),
    ('CatBoost', CatBoostClassifier(random_state=42, verbose=False))
]

# Hyperparameter tuning for the best performing models
param_grids = {
    'Logistic Regression': {'C': [0.01, 0.1, 1, 10, 100]},
    'Decision Tree': {'max_depth': [3, 5, 7, 10, None]},
    'Random Forest': {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 7, 10, None]},
    'Gradient Boosting': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]},
    'SVM': {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['linear', 'rbf']},
    'k-NN': {'n_neighbors': [3, 5, 7, 9]},
    'Neural Network (MLP)': {'hidden_layer_sizes': [(50,), (100,), (50, 50)], 'alpha': [0.0001, 0.001, 0.01]},
    'XGBoost': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]},
    'LightGBM': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]},
    'CatBoost': {'iterations': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]}
}

# Evaluate models using cross-validation and hyperparameter tuning
cv_scores = []
best_models = {}
for name, model in models:
    if name in param_grids:
        grid_search = GridSearchCV(model, param_grids[name], cv=5, scoring='accuracy')
        grid_search.fit(X_train_scaled, y_train_resampled)
        best_model = grid_search.best_estimator_
        best_models[name] = best_model
    else:
        best_model = model
        best_model.fit(X_train_scaled, y_train_resampled)
        best_models[name] = best_model

    scores = cross_val_score(best_model, X_train_scaled, y_train_resampled, cv=5, scoring='accuracy')
    cv_scores.append((name, scores.mean()))

# Print cross-validation scores
print("Cross-validation scores:")
for name, score in cv_scores:
    print(f"{name}: {score:.4f}")

# Select the best 3 models based on cross-validation scores
cv_scores.sort(key=lambda x: x[1], reverse=True)
best_model_names = [name for name, _ in cv_scores[:3]]
best_models_stacking = [(name, best_models[name]) for name in best_model_names]

# Create the stacking classifier with the best 3 models
meta_model = LogisticRegression()
stacking_model = StackingClassifier(estimators=best_models_stacking, final_estimator=meta_model)

# Train the stacking model
stacking_model.fit(X_train_scaled, y_train_resampled)

# Save the stacking model and the scaler
with open('stacking_model.pkl', 'wb') as file:
    pickle.dump(stacking_model, file)
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)

# Make predictions on the testing set with the stacking model
y_pred = stacking_model.predict(X_test_scaled)

# Print evaluation metrics for the stacking model
print("\nStacking Model")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred, average='weighted'):.4f}")
print(f"Recall: {recall_score(y_test, y_pred, average='weighted'):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred, average='weighted'):.4f}")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
import pickle
import pandas as pd

# Load the saved models
with open('stacking_model.pkl', 'rb') as file:
    loaded_stacking_model = pickle.load(file)

with open('regressor_home_goals.pkl', 'rb') as file:
    regressor_home_goals = pickle.load(file)
with open('regressor_away_goals.pkl', 'rb') as file:
    regressor_away_goals = pickle.load(file)
with open('regressor_home_corners.pkl', 'rb') as file:
    regressor_home_corners = pickle.load(file)
with open('regressor_away_corners.pkl', 'rb') as file:
    regressor_away_corners = pickle.load(file)
with open('regressor_home_yellows.pkl', 'rb') as file:
    regressor_home_yellows = pickle.load(file)
with open('regressor_away_yellows.pkl', 'rb') as file:
    regressor_away_yellows = pickle.load(file)

# Load the scaler and label encoder
with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

with open('label_encoder.pkl', 'rb') as file:
    label_encoder = pickle.load(file)

# Function to predict the outcome of a match and other stats
def predict_match_stats(home_team, away_team):
    # Check if the home team and away team have data in the dataframe
    home_team_data = df_new[df_new['HomeTeam'] == home_team]
    away_team_data = df_new[df_new['AwayTeam'] == away_team]

    if home_team_data.empty:
        print(f"Error: {home_team} does not have data in the DataFrame.")
        return None
    if away_team_data.empty:
        print(f"Error: {away_team} does not have data in the DataFrame.")
        return None

    # Prepare the input features for prediction
    home_team_features = home_team_data.iloc[-1][['HomeTeamPoints', 'HomeTeamRecentForm', 'HomeTeamGoalDifference', 'HomeTeamAvgGoalsScored', 'HomeTeamAvgGoalsConceded', 'HomeTeamLast10Wins', 'HomeTeamLast10Draws', 'HomeTeamLast10Losses']]
    away_team_features = away_team_data.iloc[-1][['AwayTeamPoints', 'AwayTeamRecentForm', 'AwayTeamGoalDifference', 'AwayTeamAvgGoalsScored', 'AwayTeamAvgGoalsConceded', 'AwayTeamLast10Wins', 'AwayTeamLast10Draws', 'AwayTeamLast10Losses']]

    # Combine the home and away team features into a single input
    input_features = pd.concat([home_team_features, away_team_features]).values.reshape(1, -1)

    # Scale the input features using the same scaler used during training
    input_features_scaled = scaler.transform(input_features)

    # Predict the match outcome
    outcome_prediction = loaded_stacking_model.predict(input_features_scaled)
    outcome_probabilities = loaded_stacking_model.predict_proba(input_features_scaled)

    # Decode the prediction
    decoded_prediction = label_encoder.inverse_transform(outcome_prediction)[0]

    # Predict goals, corners, and yellow cards
    predicted_home_goals = regressor_home_goals.predict(input_features_scaled)[0]
    predicted_away_goals = regressor_away_goals.predict(input_features_scaled)[0]
    predicted_home_corners = max(regressor_home_corners.predict(input_features_scaled)[0], 0)
    predicted_away_corners = max(regressor_away_corners.predict(input_features_scaled)[0], 0)
    predicted_home_yellows = max(regressor_home_yellows.predict(input_features_scaled)[0], 0)
    predicted_away_yellows = max(regressor_away_yellows.predict(input_features_scaled)[0], 0)

    # Convert goals, corners, and yellow cards to over/under format
    def convert_to_over_under(value, thresholds):
        for threshold in thresholds:
            if value <= threshold:
                return f"Under {threshold}"
        return f"Over {thresholds[-1]}"

    goals_thresholds = [1.5, 2.5, 3.5, 4.5]
    corners_thresholds = [5, 10, 15]
    yellows_thresholds = [1, 2, 3, 4]

    home_goals_over_under = convert_to_over_under(predicted_home_goals, goals_thresholds)
    away_goals_over_under = convert_to_over_under(predicted_away_goals, goals_thresholds)
    home_corners_over_under = convert_to_over_under(predicted_home_corners, corners_thresholds)
    away_corners_over_under = convert_to_over_under(predicted_away_corners, corners_thresholds)
    home_yellows_over_under = convert_to_over_under(predicted_home_yellows, yellows_thresholds)
    away_yellows_over_under = convert_to_over_under(predicted_away_yellows, yellows_thresholds)

    # Predict total goals
    total_goals = predicted_home_goals + predicted_away_goals
    total_goals_over_under = convert_to_over_under(total_goals, goals_thresholds)

    # Format the probabilities
    probabilities = {
        "Home Win": outcome_probabilities[0][0] * 100,
        "Draw": outcome_probabilities[0][1] * 100,
        "Away Win": outcome_probabilities[0][2] * 100
    }

    return {
        "Predicted Outcome": "Home Win" if decoded_prediction == "H" else ("Draw" if decoded_prediction == "D" else "Away Win"),
        "Probabilities": probabilities,
        "Predicted Goals": {
            "Home": home_goals_over_under,
            "Away": away_goals_over_under
        },
        "Predicted Corners": {
            "Home": home_corners_over_under,
            "Away": away_corners_over_under
        },
        "Predicted Yellow Cards": {
            "Home": home_yellows_over_under,
            "Away": away_yellows_over_under
        },
        "Predicted Total Goals": total_goals_over_under
    }

# Example usage
home_team = 'Liverpool'
away_team = 'Everton'
predictions = predict_match_stats(home_team, away_team)

if predictions:
    print(f"Predicted outcome for {home_team} vs {away_team}: {predictions['Predicted Outcome']}")
    print(f"Probabilities: Home Win - {predictions['Probabilities']['Home Win']:.2f}%, Draw - {predictions['Probabilities']['Draw']:.2f}%, Away Win - {predictions['Probabilities']['Away Win']:.2f}%")
    print(f"Predicted Goals: {home_team} - {predictions['Predicted Goals']['Home']}, {away_team} - {predictions['Predicted Goals']['Away']}")
    print(f"Predicted Corners: {home_team}  - {predictions['Predicted Corners']['Home']}, {away_team} - {predictions['Predicted Corners']['Away']}")
    print(f"Predicted Yellow Cards: {home_team}  - {predictions['Predicted Yellow Cards']['Home']}, {away_team} - {predictions['Predicted Yellow Cards']['Away']}")
    print(f"Predicted Total Goals: {predictions['Predicted Total Goals']}")
```
